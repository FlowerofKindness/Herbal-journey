## Data collection and knowledge representation

### Subtask:
Gather information on various herbs, their medicinal properties, traditional uses, and associated plant spirit principles. Structure this information in a way that can be easily accessed and utilized by the application (e.g., a database or knowledge graph).


**Reasoning**:

A good way to represent this initial data collection and organization is using a pandas DataFrame, which is flexible and can be easily converted to other storage formats later if needed. I will create a sample DataFrame with a few herbs and relevant fields as a starting point to demonstrate the structure.



```python
import pandas as pd

# Create a sample DataFrame to structure the herb information
data = {
    'herb_name': ['Chamomile', 'Peppermint', 'Lavender', 'Echinacea'],
    'common_names': ['German Chamomile', 'Mint', 'True Lavender', 'Purple Coneflower'],
    'medicinal_properties': ['calming, anti-inflammatory', 'digestive aid, stimulant', 'calming, antiseptic', 'immune booster'],
    'traditional_uses': ['anxiety, insomnia, indigestion', 'indigestion, headaches', 'anxiety, sleep, skin irritation', 'cold, flu'],
    'associated_symptoms': ['anxiety, insomnia, indigestion', 'indigestion, headache', 'anxiety, insomnia, skin irritation', 'cold, flu symptoms'],
    'plant_spirit_principles': ['calm, peace', 'invigoration, clarity', 'tranquility, healing', 'strength, resilience']
}

df_herbs = pd.DataFrame(data)

# Display the created DataFrame
display(df_herbs)
```

## Symptom input and mapping

### Subtask:
Design a system for users to input their symptoms or desired outcomes. Develop a method to map these inputs to relevant herbs based on their properties and principles.


**Reasoning**:
Define a function to find herbs based on user input symptoms by iterating through the DataFrame and checking for symptom matches.



```python
def find_herbs_by_symptom(symptom, df):
    """
    Finds herbs based on user-input symptoms.

    Args:
        symptom: The symptom string provided by the user.
        df: The DataFrame containing herb information.

    Returns:
        A DataFrame containing rows of herbs that match the symptom.
    """
    matching_herbs = df[df['associated_symptoms'].str.contains(symptom, case=False, na=False)]
    return matching_herbs

# Example usage (for testing purposes within the cell)
# Replace 'insomnia' with a symptom you want to test
# found_herbs = find_herbs_by_symptom('insomnia', df_herbs)
# display(found_herbs)
```
